```python
from pyspark.sql import SparkSession
spark = SparkSession.builder \
    .master("local") \
    .appName("Nome do Aplicativo") \
    .config("spark.some.config.option", "some-value") \
    .getOrCreate()
```

```python
sc = spark.sparkContext
```

```python
arquivo_raw = sc.textFile('/home/heliton/Documentos/repos/aws_cloud_data_engineer_compass_uol/README.md')
# arquivo_raw = spark.read.text('/home/heliton/Documentos/repos/novo_aws_cloud_data_engineer_compass_uol/README.md')
```

```python
arquivo_raw.collect()
```

```python
import re
palavras = arquivo_raw.flatMap(lambda line: re.split('\W+', line.lower().strip()))
```

```python
palavras = palavras.filter(lambda x: len(x) >= 1)
```

```python
palavras = palavras.map(lambda w: (w,1))
```

```python
from operator import add
palavras = palavras.reduceByKey(add)
```

```python
palavras.take(10)
```

```python
palavras_order = palavras.map(lambda x:(x[1],x[0])).sortByKey(False)
```

```python
palavras_order.take(10)
```

```python
palavras_order.collect()
```
